Task - 3 - Outlier Detection:

```python
#Compute IQR for Score
Q1_score = south_asian_df["score"].quantile(0.25)
Q3_score = south_asian_df["score"].quantile(0.75)
IQR_score = Q3_score - Q1_score

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Compute IQR for GDP per capita
Q1_gdp = south_asian_df["Log GDP per capita"].quantile(0.25)
Q3_gdp = south_asian_df["Log GDP per capita"].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify Outliers
south_asian_df["Score Outlier"] = (
    (south_asian_df["score"] < lower_bound_score) |
    (south_asian_df["score"] > upper_bound_score)
)

south_asian_df["GDP Outlier"] = (
    (south_asian_df["Log GDP per capita"] < lower_bound_gdp) |
    (south_asian_df["Log GDP per capita"] > upper_bound_gdp)
)

# Countries that are outliers in either Score OR GDP
south_asian_df["Any Outlier"] = south_asian_df["Score Outlier"] | south_asian_df["GDP Outlier"]

outliers = south_asian_df[south_asian_df["Any Outlier"]]
print("\nOutlier Countries:")
print(outliers[["Country name", "score", "Log GDP per capita"]])

#Scatter Plot Highlighting Outliers
plt.figure(figsize=(10, 6))

# normal points
normal = south_asian_df[south_asian_df["Any Outlier"] == False]
plt.scatter(normal["Log GDP per capita"], normal["score"], label="Normal", s=80)

# outlier points in red
plt.scatter(outliers["Log GDP per capita"], outliers["score"],
            color="red", label="Outliers", s=120)

# add labels to outliers
for i, row in outliers.iterrows():
    plt.text(row["Log GDP per capita"], row["score"], row["Country name"])

plt.xlabel("GDP per Capita (Logged)")
plt.ylabel("Happiness Score (Ladder Score)")
plt.title("Outlier Detection in South Asian Countries")
plt.legend()
plt.show()
```

5. Correlation Analysis:

```python
def create_mock_data(n=50, region_name="Region"):
    data = {
        'score': np.random.uniform(3, 8, n),
        'Freedom to make life choices': np.random.uniform(0.4, 0.9, n),
        'Generosity': np.random.uniform(-0.1, 0.5, n),
        'Region': region_name
    }
    return pd.DataFrame(data)

south_df = create_mock_data(50, "South Asia")
middle_df = create_mock_data(50, "Middle East")

# CALCULATE CORRELATIONS
def calculate_correlations(df, region_name):
    corr_freedom = df['score'].corr(df['Freedom to make life choices'])
    corr_generosity = df['score'].corr(df['Generosity'])
    print(f"--- {region_name} ---")
    print(f"Pearson Correlation (Score vs Freedom):    {corr_freedom:.3f}")
    print(f"Pearson Correlation (Score vs Generosity): {corr_generosity:.3f}")
    print("-" * 30)
calculate_correlations(south_df, "South Asia")
calculate_correlations(middle_df, "Middle East")

# PLOTTING FUNCTION
def plot_trend(df, x_col, y_col, title, color, ax):
    """
    Plots a scatter plot with a regression line using seaborn on a specific axis.
    """
    # regplot creates both the scatter and the linear regression line
    sns.regplot(data=df, x=x_col, y=y_col, color=color, ax=ax,line_kws={"color": "red", "linestyle": "--"},
        scatter_kws={"alpha": 0.7})

    ax.set_title(title, fontsize=12)
    ax.set_xlabel(x_col, fontsize=10)
    ax.set_ylabel('Happiness Score', fontsize=10)
    ax.grid(True, linestyle='--', alpha=0.6)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

plot_trend(middle_df, 'Freedom to make life choices', 'score', 'Freedom vs Score (Middle East)', 'blue',ax=axes[0, 0])

plot_trend(middle_df, 'Generosity', 'score', 'Generosity vs Score (Middle East)', 'blue', ax=axes[0, 1])

plot_trend(south_df, 'Freedom to make life choices', 'score', 'Freedom vs Score (South Asia)', 'green',ax=axes[1, 0])

plot_trend(south_df, 'Generosity', 'score', 'Generosity vs Score (South Asia)', 'green',ax=axes[1, 1])

plt.tight_layout()
plt.show()
```

6. Outlier Detection:

```python
#Function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

#Function to flag outliers in DataFrame
def flag_outliers(df, score_col="score", gdp_col="Log GDP per capita"):
    df["Score Outlier"] = (
        (df[score_col] < df[score_col].quantile(0.25) - 1.5*(df[score_col].quantile(0.75)-df[score_col].quantile(0.25))) |
        (df[score_col] > df[score_col].quantile(0.75) + 1.5*(df[score_col].quantile(0.75)-df[score_col].quantile(0.25)))
    )

    df["GDP Outlier"] = (
        (df[gdp_col] < df[gdp_col].quantile(0.25) - 1.5*(df[gdp_col].quantile(0.75)-df[gdp_col].quantile(0.25))) |
        (df[gdp_col] > df[gdp_col].quantile(0.75) + 1.5*(df[gdp_col].quantile(0.75)-df[gdp_col].quantile(0.25)))
    )

    df["Any Outlier"] = df["Score Outlier"] | df["GDP Outlier"]
    return df

# Flag outliers - Ensuring original dataframes are used
south_asian_df = flag_outliers(south_asian_df)
middle_east_df = flag_outliers(middle_east_df)

# Prepare for plotting
regions = {"South Asia": south_asian_df, "Middle East": middle_east_df}

plt.figure(figsize=(16, 6))

for i, (region_name, df) in enumerate(regions.items(), 1):
    plt.subplot(1, 2, i)
    normal = df[df["Any Outlier"] == False]
    outliers = df[df["Any Outlier"] == True]

    plt.scatter(normal["Log GDP per capita"], normal["score"], label="Normal", s=80, alpha=0.7)
    plt.scatter(outliers["Log GDP per capita"], outliers["score"], color="red", label="Outliers", s=120, alpha=0.9, edgecolors='black')

    for _, row in outliers.iterrows():
        plt.text(row["Log GDP per capita"] + 0.02, row["score"], row["Country name"], fontsize=9)

    plt.xlabel("Log GDP per Capita")
    plt.ylabel("Happiness Score")
    plt.title(f"{region_name}: Score vs GDP Outliers")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

## Implications of Outliers for South Asia and Middle East

**South Asia Region:**

* **Identified Outlier:** Afghanistan is a clear outlier, with much lower happiness scores and Log GDP per capita than all other South Asian countries.
* **Implications:** This outlier skews regional averages and distorts overall trends, reflecting Afghanistans unique challenges such as conflict and economic instability. It highlights that South Asias general patterns cannot be fully understood without considering Afghanistan separately.

**Middle East Region:**

* **Identified Outliers:** No major outliers were found in the Middle East based on the 1.5  IQR rule.
* **Implications:** The absence of outliers indicates a more consistent distribution of happiness and economic levels across the region, making regional trends more stable and representative compared to South Asia.
